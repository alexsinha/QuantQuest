```python
columns=['year','month']
```

```python
#handling categorical columns
def category_multcols(multcolumns):
    df_final=all_data
    i=0
    for fields in multcolumns:
        
        print(fields)
        df1=pd.get_dummies(all_data[fields],drop_first=False)
        
        all_data.drop([fields],axis=1,inplace=True)
        if i==0:
            df_final=df1.copy()
        else:
            df_final=pd.concat([df_final,df1],axis=1)
        i=i+1
       
    df_final=pd.concat([all_data,df_final],axis=1)
        
    return df_final
```

```python
all_data=category_multcols(columns)
```

```python
#coverting into matrix
X=np.matrix(all_data)
y=np.matrix(revenue)
X.shape,y.shape
```

```python
#Building Model
import xgboost
regressor=xgboost.XGBRegressor()
from sklearn.model_selection import RandomizedSearchCV
```

```python
## Hyper Parameter Optimization
n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
booster=['gbtree','gblinear']
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]
base_score=[0.25,0.5,0.75,1]
# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }
```

```python
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
```

```python
random_cv.fit(X,y)
```

```python
random_cv.score(X, y)
```
